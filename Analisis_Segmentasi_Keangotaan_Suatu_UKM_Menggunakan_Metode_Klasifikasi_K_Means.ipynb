##**Analisis Segmentasi Keangotaan Suatu UKM Menggunakan Metode Klasifikasi K-Means**

```python
#impor library

import numpy as np #linear algebra
import pandas as pd #proses data
import matplotlib.pyplot as plt #Visualisasi Data
```

```python
url = 'https://raw.githubusercontent.com/Ellayuninf/Riset-Informatika/main/data-mhs.csv'
dataset = pd.read_csv(url)

dataset.head(5)
```

```python
print(dataset.columns)
print(dataset.head())
print(dataset.info())
print(dataset.describe())
```

```python
# Menghitung frekuensi untuk setiap gender
gender_counts = dataset['Jenis_Kelamin'].value_counts()

# Menentukan warna berdasarkan label
colors = ['pink', 'blue']

# Membuat bar plot
plt.figure(figsize=(10, 5))
plt.bar(gender_counts.index, gender_counts.values, color=colors)
plt.title('Perbandingan Jumlah Pengunjung Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis_Kelamin')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Menemukan modus dari kolom "Jurusan"
modus_jurusan = dataset['Jurusan'].mode()

# Menampilkan hasil
print("Modus Jurusan:", modus_jurusan)
```
